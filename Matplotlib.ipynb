```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
plt.plot ([1,2,3],[5,7,4])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

```python
plt.style.use('classic')
import numpy as np
x = [1,2,3]
y = [5,7,4]
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Standardowy wykres")
plt.show()
```

```python
x = [1,2,3]
y = [5,7,4]
x2 = [1,2,3]
y2 = [4,6,8]
x3 = [3,4,5]
y3 = [6,7,8]
plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y,label = 'Opis 1')
plt.plot(x2,y2,label = 'Opis 2')
plt.plot(x3,y3,label = 'Opis 3')
plt.title("Tytul")
plt.legend()
plt.show()
```

```python
import numpy as np
x = np.linspace(0,2.0*np.pi,101)
y= np.sin(x)

plt.plot(x,y, label= 'Sine Curve',color ='b')
plt.xlabel("Kąt w radianach")
plt.ylabel("Wielkość")
plt.title("Tytuł")
plt.grid() 
xnumbers = np.linspace(0,7,10)
ynumbers = np.linspace(-1,1,10)
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.axis([0,6.3,-1.2,1.2])
plt.legend()
plt.show()
```

```python
x = np.linspace(0,2.0*np.pi,101)
y= np.sin(x)
z = np.cos(x)
plt.plot(x,y,'b',x,z,'r')
plt.xlabel("Kąt w radianach")
plt.ylabel("Wielkość")
plt.title("Funkcja sin/cos")
plt.grid() 
xnumbers = np.linspace(0,7,10)
ynumbers = np.linspace(-1,1,10)
plt.xticks(xnumbers)
plt.yticks(ynumbers)
plt.axis([0,6.3,-1.2,1.2])
plt.legend(['sin','cos'])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,2]
x2 = [1,4,6,8,4]
y2 =[3,2,1,6,9]
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Tytuł")
plt.bar(x,y, label='Dane A', color ='r')
plt.bar(x2,y2, label='Dane B',color = 'b')
plt.legend()
plt.show()
```
